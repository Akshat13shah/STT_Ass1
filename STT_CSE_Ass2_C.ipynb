```python
from pydriller import Repository
import csv

repo_path = r"https://github.com/activeloopai/deeplake"

# Expanded bug-related keywords from your list
bug_keywords = [
    "fixed", "bug", "fixes", "fix ", " fix", " fixed", " fixes", "crash",
    "solves", "resolves", " resolves ", " issue", "issue", "regression",
    "fall back", "assertion", "coverity", "reproducible", "stack-wanted",
    "steps-wanted", "testcase", "failure", "fail", "npe", " npe", "except",
    "broken", "differential testing", "error", "hang", " test fix",
    "steps to reproduce", "crash", "failure", "leak", "stack trace",
    "heap overflow", "freez", "problem", " overflow", "overflow", "avoid",
    "workaround", "break", "stop"
]

# CSV file output path
output_csv = "bug_fixes.csv"

# Open CSV and write header
with open(output_csv, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow([
        "Hash", "Message", "Parent Hashes",
        "Is Merge Commit", "Modified Files"
    ])

    for commit in Repository(repo_path).traverse_commits():
        msg_lower = commit.msg.lower()

        # Check if commit message contains any bug keyword
        if any(keyword in msg_lower for keyword in bug_keywords):
            writer.writerow([
                commit.hash,
                commit.msg.strip(),
                ";".join(commit.parents),
                commit.merge,
                ";".join(mod.filename for mod in commit.modified_files if mod.filename)
            ])

print(f"Bug-fixing commits saved to {output_csv}")
```
